# Stimuli Generation

Use the `Stimuli()` class to come up with the stimuli of your circuit and finally save it in a file that is called the same as this notebook but with `*.json` extension.

```python
from opens import Stimuli
import numpy as np
import matplotlib.pyplot as plt

stimuli = Stimuli()

stimuli["t"] = np.linspace(0, 100e-3, 200)
stimuli["L"] = 320 * np.sin(2*np.pi*50*stimuli["t"]) # this will create a pwl source with these discrete 
stimuli["N"] = 0

plt.plot(stimuli["t"], stimuli["L"])
```

```python
stimuli.save("stimuli.json")
```
