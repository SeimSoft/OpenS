# Design Script

Use the `DesignPoints()` class to come up with your dimensions for your circuit and finally save it in a file that is called the same as this notebook but with `*.json` extension.

```python
from opens import DesignPoints

dps = DesignPoints()

dps["R1.R [Ohm]"] = 10e3

dps
```
